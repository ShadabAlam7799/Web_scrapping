```python
def extract_categories(baseurl):
    categories = {}
    r = requests.get(baseurl)
    soup = BeautifulSoup(r.content)
    categories_list = soup.find("ul", class_="nav-list").ul.find_all("li")
    for elm in categories_list:
        categories.update({elm.text.strip():baseurl+elm.a["href"]})
    return categories
```

```python
cat_dict = extract_categories(baseurl)
```

```python
def extract_categor(baseurl):
    categor = {"cat":[], "link":[]}
    r = requests.get(baseurl)
    soup = BeautifulSoup(r.content)
    categories_list = soup.find("ul", class_="nav-list").ul.find_all("li")
    for elm in categories_list:
        categor["cat"].append(elm.text.strip())
        categor["link"].append(baseurl+elm.a["href"])
    return categor
```

```python
cat_d = extract_categor(baseurl)
```

```python
url = "https://books.toscrape.com/catalogue/category/books/mystery_3/index.html"
```

```python
book_page_list = soup.find("ol", class_="row").find_all("li")
```

```python
book_page_list
```

```python
temp_url = book_page_list[0].a["href"].split("/")[3:]
```

```python
temp_url
```

```python
"/".join(temp_url)
```

```python
baseurl + "catalogue/" + "/".join(temp_url)
```

```python
data = {
    "title":[],
    "book_url":[],
    "price_with_tax":[],
    "availability":[]
}
```

```python
for elm in book_page_list:
    href_split = elm.a["href"].split("/")[3:]
    book_page_link = baseurl + "catalogue/" + "/".join(href_split)
    
    ## send a request to the book page link to get to it
    book_page_res = requests.get(book_page_link)
    
    ## create the soup object to get the HTML
    soup = BeautifulSoup(book_page_res.content)
    
    data["book_url"].append(book_page_link)
    
    title = soup.find("div", class_="product_main").h1.text.strip()
    data["title"].append(title)
    
    rows = soup.find("table", class_="table-striped").find_all("tr")
    data["price_with_tax"].append(rows[3].find("td").text.strip())
    data["availability"].append(rows[5].find("td").text.strip())
```
